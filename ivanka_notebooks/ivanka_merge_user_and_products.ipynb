```python
import pandas as pd

# Завантаження даних
orders = pd.read_csv('/Users/ivanna/Documents/University/semester_5/PROJECT/new_data/orders_final.csv')  # Замовлення з user_id
order_products = pd.read_csv('/Users/ivanna/Documents/University/semester_5/PROJECT/new_data/order_products_prior_final.csv')  # Продукти в замовленнях

# Об'єднання таблиць за order_id
merged_data = pd.merge(order_products, orders[['order_id', 'user_id']], on='order_id', how='inner')

# Переглядаємо результат
print(merged_data.head())
```

```python
# Групуємо дані за user_id та order_id, збираємо продукти для кожного замовлення в список
grouped_orders = (
    merged_data.groupby(['user_id', 'order_id'])['product_id']
    .apply(list)
    .reset_index()
)

grouped_orders.to_csv('user_products_grouped.csv', index=False)

# Перевіряємо структуру згрупованих даних
print(grouped_orders.head())
```

```python
import pandas as pd
from itertools import combinations

# File paths
input_file = 'user_products_grouped.csv'
output_file = 'order_product_pairs.csv'

# Define the chunk size
chunk_size = 100  # Number of orders per chunk

# Write the header to the output file
with open(output_file, 'w') as f_out:
    f_out.write('user_id,order_id,product_1,product_2\n')

print("Processing chunks...")

# Read the input CSV in chunks
chunks = pd.read_csv(input_file, chunksize=chunk_size)

for chunk_num, chunk in enumerate(chunks, start=1):
    print(f"Processing chunk {chunk_num}...")

    # Open the output file in append mode to add results
    with open(output_file, 'a') as f_out:
        # Iterate over each order in the chunk
        for idx, row in chunk.iterrows():
            user_id = row['user_id']
            order_id = row['order_id']

            # Convert the product list from string to a list (if needed)
            products = eval(row['product_id'])  # Ensure it’s a list

            # Generate combinations of 2 products
            pairs = combinations(products, 2)

            # Sort each pair and write to the output file
            for pair in pairs:
                sorted_pair = tuple(sorted(pair))  # Ensure (a, b) == (b, a)
                f_out.write(f"{user_id},{order_id},{sorted_pair[0]},{sorted_pair[1]}\n")

print("Processing complete! Results saved to 'order_product_pairs.csv'.")
```

```python
import pandas as pd

# Завантаження CSV з парами продуктів
pairs_df = pd.read_csv('order_product_pairs.csv')

# Підрахунок частоти кожної унікальної пари
pair_counts = (
    pairs_df.groupby(['product_1', 'product_2'])
    .size()
    .reset_index(name='count')
)

# Сортуємо пари за кількістю у спадному порядку
pair_counts = pair_counts.sort_values(by='count', ascending=False)

# Зберігаємо результати у новий CSV
pair_counts.to_csv('unique_pair_counts.csv', index=False)

# Виводимо топ-10 найпопулярніших пар
print(pair_counts.head(10))
```
